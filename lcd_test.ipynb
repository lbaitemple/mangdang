```python
from PIL import Image, ImageDraw, ImageFont
from MangDang.LCD.ST7789 import ST7789

hello_text = "My name is Pupper!"

disp = ST7789()
disp.begin()

# background image
image = Image.open('/var/lib/mini_pupper_bsp/test.png')
# resize the image
image.resize((320, 240))
# choose a text font
font = ImageFont.truetype('/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf', 30)
# draw the image with text with font
draw = ImageDraw.Draw(image)
draw.text((10, 95), hello_text, font=font, fill="#000000", spacing=0, align='left')
# display
disp.display(image)
```
